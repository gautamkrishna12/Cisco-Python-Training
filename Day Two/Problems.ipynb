```python
sentance=input('Enter the senetence :')
print(f'Text entered : {sentance}')
list_sent=sentance.split()
tuple_sent=tuple(list_sent)
print(f'List : {list_sent}')
print(f'Tuple : {tuple_sent}')

with open('Problem 1.txt',mode='w') as file:
    file.write(f'Tuple : {tuple_sent}\n')
    file.write(f'List : {list_sent}')
    print('File opened and writed')

```

```python
data_list=list(map(int,input('Enter integers seperated by spaces : ').split()))
print(f'Enter numbers are : {data_list}')
Sum,avg=0,0
for i in data_list:
    Sum+=i
avg=Sum/len(data_list)
with open('Problem 2.txt',mode='w') as file:
    file.write(f'Numbers entered : {data_list}\n')
    file.write(f'Sum : {Sum}\n')
    file.write(f'Average : {avg}\n')
```

```python
with open('Problem 2.txt',mode='r') as file:
    for i in file.readlines():
        print(i)
```
