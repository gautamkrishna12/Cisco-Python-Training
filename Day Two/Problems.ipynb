```python
#Problem 2
```

```python
data_list=list(map(int,input('Enter integers seperated by spaces : ').split()))
print(f'Enter numbers are : {data_list}')
Sum,avg=0,0
for i in data_list:
    Sum+=i
avg=Sum/len(data_list)
with open('numbers_data.txt',mode='w') as file:
    file.write(f'Numbers entered : {data_list}\n')
    file.write(f'Sum : {Sum}\n')
    file.write(f'Average : {avg}\n')
```

```python
with open('numbers_data.txt',mode='r') as file:
    for i in file.readlines():
        print(i)
```

```python
#Problem 3
```

```python
sentence=input('Enter the Sentence : ')
words=sentence.split()
word_tuple=[]
for i in words:
    word_tuple.append(i.upper())
word_tuple=tuple(word_tuple)
with open('sentence_data.txt',mode='w') as file:
    file.write(f'Tuple : {word_tuple}\n')
    file.write(f'List : {words}')
    print('File opened and writed')
with open('sentence_data.txt',mode='r') as file:
    print('Printing file data')
    for i in file.readlines():
        print(i)
```

```python
Names=input('Enter the names (Seperated by space) :').split()
Names.sort()
with open('names_data.txt',mode='w') as file:
    file.write(f'Tuple : {tuple(Names)}\n')
    print('File opened and writed')
```

```python
with open('names_data.txt',mode='r') as file:
    print('Printing file data')
    for i in file.readlines():
        print(i)
```

```python
data_list=list(map(int,input('Enter the numbers seperated by spaces : ').split()))
Max=max(data_list)
Min=min(data_list)
with open('minmax_data.txt',mode='w') as file:
    file.write(f'Numbers entered : {data_list}\n')
    file.write(f'Maximum Value : {Max}\n')
    file.write(f'Minimum Value : {Min}\n')
```

```python
with open('minmax_data.txt',mode='r') as file:
    for i in file.readlines():
        print(i)
```
