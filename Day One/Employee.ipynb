```python
Employees={}

def AddEmployee():
    emp_id=int(input('Enter the employee id :'))
    if emp_id in Employees:
        return f'Employee with employee id {emp_id} already exists'
    else:
        name=input('Enter the name :')
        department=input('Enter the department :')
        salary=int(input('Enter salary :'))
        Employees[emp_id]=(name,department,salary)
```

```python
def Search_Employee():
    emp_id=int(input('Enter the employee id :')) 
    if emp_id in Employees:
        print(f'Employee ID : {emp_id}')
        print(f'Name : {Employees[emp_id][0]}')
        print(f'Department : {Employees[emp_id][1]}')
        print(f'Salary : {Employees[emp_id][2]}')
    else:
        return f'Employee with Employee ID {emp_id} not exists.'
```

```python
def Update_Employee_Salary():
    emp_id=int(input('Enter the employee id :')) 
    if emp_id in Employees:
        new_salary=int(input('Enter the new salary :'))
        Employees[emp_id]=(Employees[emp_id][0],Employees[emp_id][1],new_salary)
    else:
        return f'Employee with Employee ID {emp_id} not exists.'
```

```python
def Remove_Employee():
    emp_id=int(input('Enter the employee id :')) 
    if emp_id in Employees:
        del Employees[emp_id]
        return 'Employee is removed'
    else:
        return f'Employee with Employee ID {emp_id} not exists.'
```

```python
class Employee():
    def __init__(self,emp_id,name,department,salary):
        self.Emp_id=emp_id
        self.Name=name
        self.Department=department
        self.Salary=salary

    def Update_Salary(self):
        new_salary=int(input('Enter the new salary'))
        self.Salary=new_salary
        return 'Salary updated successfully'
    
    def Change_Department(self):
        new_dept=input('Enter the new department')
        self.Department=new_dept
        return 'Department updated Successfully'

    def Employee_Details(self):
        print(f'Employee ID : {self.Emp_id}')
        print(f'Name : {self.Name}')
        print(f'Department : {self.Department}')
        print(f'Salary : {self.Salary}')
```

```python
def Add_NewEmployee():
    emp_id=int(input('Enter the employee id :'))
    for i in Employees:
        if i.Emp_id==emp_id:
            return f'Employee with employee id {emp_id} already exists'
    name=input('Enter the name :')
    department=input('Enter the department :')
    salary=int(input('Enter salary :'))
    emp=Employee(emp_id,name,department,salary)
    Employees.append(emp)
    return 'Employee added successfully.'
```

```python
def Search_Employee():
    emp_id=int(input('Enter the employee id :')) 
    for i in Employees:
        if i.Emp_id==emp_id:
            print("\n")
            return i.Employee_Details()
            print("\n")
    
    return f'Employee with Employee ID {emp_id} not exists.'
```

```python
def Remove_Employee():
    emp_id=int(input('Enter the employee id :')) 
    for i in Employees:
        if i.Emp_id==emp_id:
            Employees.pop(Employees.index(i))
            del i
            return 'Employee is removed successfully'
    
    return f'Employee with Employee ID {emp_id} not exists.'
```

```python
def Update_Salary():
    emp_id=int(input('Enter the employee id :')) 
    for i in Employees:
        if i.Emp_id==emp_id:
            new_salary=int(input('Enter the new salary :'))
            i.Salary=new_salary
            return 'Salary Updated successfully'
    return f'Employee with Employee ID {emp_id} not exists.'
```

```python
while True:
    print("\n****** Employee Management System ******")
    print("1. Add New Employee")
    print("2. Search Employees")
    print("3. Update Employee Salary")
    print("4. Remove Employee")
    print("5. Exit\n")
    choice=int(input('Enter your Choice: '))

    if choice==1:
        Add_Employee()
    elif choice==2:
        Search_Employee()
    elif choice==3:
        Update_Salary()
    elif choice==4:
        Remove_Employee()
    elif choice==5:
        print("***** Exiting the progra. Thank You ! *****")
        break
    else:
        print('Enter a valid choice.')
    
```
